```python
import math

import pandas as pd

def count_c(*, n: int, k: int) -> float:
  c = math.factorial(n) / (math.factorial(k) * math.factorial(n-k))
  return c


def count_p(*, x: int, y: int) -> float:
  p_x = count_c(k=x, n=8) * (1/6) ** x * (5/6) ** (8-x)
  p_y = count_c(k=y, n=11) * (1/2) ** 11
  return p_x * p_y


p = {}
max_p = (0, 0, -1) # x, y, probability
for y in range(12):
  p_ = {}
  for x in range(9):
    p_xy = count_p(x=x, y=y)
    p_[x] = p_xy
    if p_xy > max_p[2]:
      max_p = (x, y, p_xy)
    elif p_xy == max_p[2]:
      print(x, y, 'вероятность также равна максимальной на текущей момент!')
  p[y] = p_

df = pd.DataFrame(p)
df
```
